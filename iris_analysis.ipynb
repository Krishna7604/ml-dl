```python
fig,ax=plt.subplots(4,4,figsize=(20,20))
plt.suptitle("iris pairplot")
for i in range(4):
    for j in range(4):
        if i==j:
            ax[i,j].hist(iris.data[:,i])
        else:
            ax[i,j].scatter(iris.data[:,j],iris.data[:,i],c=iris.target)
            ax[i,j].set_xticks(())
            ax[i,j].set_yticks(())
        if i==3:
            ax[i,j].set_xlabel(iris.feature_names[j])
        if j==0:
            ax[i,j].set_ylabel(iris.feature_names[i])
            
```

```python
# spliting
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(iris.data,iris.target,random_state=0)
```

```python
print("dataset shepe:",iris.data.shape)
print("train dataset shape:",x_train.shape)
print("test dataset shape:",x_test.shape)
```

```python
#model selection
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=1)
```

```python
#model training
model.fit(x_train,y_train)
```

```python
# predict test data
y_pred=model.predict(x_test)
y_pred
```

```python
print(sum(y_pred==y_test),"/",y_test.shape[0], "are correct")
print("score :",model.score(x_test,y_test))
```

```python
import random
r=random.randint(1,150)
test_pred=model.predict([iris.data[r]])
print("input=",iris.data[r])
print("prediction=",iris.target_names[test_pred][0])
print("original=",iris.target_names[iris.target[r]])
```
